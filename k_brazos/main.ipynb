## Descripción

Este proyecto aborda el problema del Bandido Multi-Brazo (Multi-Armed Bandit). El objetivo es estudiar y comparar diferentes estrategias de toma de decisiones que permitan maximizar la recompensa acumulada equilibrando adecuadamente la exploración de nuevas acciones y la explotación del conocimiento adquirido. Se implementan y evalúan varias familias de algoritmos:

* **ε-Greedy**: estrategia basada en la selección del brazo con mayor recompensa estimada, incorporando un mecanismo de exploración aleatoria controlada.
* **Upper Confidence Bound (UCB)**: método que utiliza la confianza para favorecer la exploración de acciones con mayor incertidumbre.
* **Métodos de Ascenso de Gradiente**: enfoque basado el ascenso del gradiente para estimar los valores de las acciones.

Los algoritmos se comparan utilizando diferentes métricas de rendimiento con el fin de analizar su comportamiento y eficiencia:

* **Recompensa promedio acumulada**.
* **Porcentaje de selección del brazo óptimo** a lo largo del aprendizaje.
* **Histograma de selección de brazos y recompensa media**, mostrando el número de veces que cada brazo ha sido seleccionado junto con la recompensa promedio obtenida por dicho brazo.
* **Regret acumulado**, utilizado para cuantificar la diferencia respecto a una estrategia óptima.

Los experimentos se realizan sobre distintas configuraciones del problema:

* **Bernoulli**: cada acción genera una recompensa binaria (éxito o fracaso) según una probabilidad fija.
* **Binomial**: la recompensa se obtiene a partir de múltiples ensayos independientes con probabilidad de éxito constante.
* **Normal**: las recompensas se generan a partir de una distribución continua caracterizada por una media y una desviación estándar.
