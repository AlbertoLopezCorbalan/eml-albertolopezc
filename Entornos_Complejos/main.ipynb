
## Descripción

Este proyecto estudia y compara diferentes algoritmos de aprendizaje por refuerzo en distintos entornos proporcionados por **Gymnasium**. Se distinguen dos métodos:

- **Métodos tabulares**, aplicados a entornos discretos como *Taxi-v3*. https://gymnasium.farama.org/environments/toy_text/taxi/
- **Métodos de aproximación**, aplicados a entornos continuos como *LunarLander-v3*. https://gymnasium.farama.org/environments/box2d/lunar_lander/
